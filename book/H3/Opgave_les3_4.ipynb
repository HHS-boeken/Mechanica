De functie voor de elastische lijn (doorbuigings) kan gevonden worden door gebruik te maken van;

$EI \dfrac{d^4v}{dx^4}= 0$ en deze twee keer te integreren.

Voor snede I - tussen $[0 < x < 5]$ geldt;


\begin{align}
EI\dfrac{d^4v}{dx^4} &= 0 \\
EI\dfrac{d^3v}{dx^3} &= C_1 \\
EI\dfrac{d^2v}{dx^2} &= C_1x + C_2 \\
EI\dfrac{dv}{dx} &= \dfrac{1}{2}C_1x^2 + C_2x + C_3 \\
EIv &= \dfrac{1}{6}C_1x^3 + \dfrac{1}{2}C_2x^2 + C_3x +C_4
\end{align}


De integratie constanten $C_1$, $C_2$, $C_3$ en $C_4$ kunnen gevonden worden door gebruik te maken van de volgende randvoorwaardes.

Voor snede I - tussen $[0 < x < 5]$ geldt;

Bij oplegging $A$ geldt;

$x=0, v=0$


\begin{align}
    EIv &= \dfrac{1}{6}C_1x^3 + \dfrac{1}{2}C_2x^2 + C_3x +C_4 \\
    EI\cdot 0 &= \dfrac{1}{6}C_1 \cdot 0^3 + \dfrac{1}{2}C_2 \cdot 0^2 + C_3 \cdot 0 +C_4  \\
    0 &= 0 + 0 +0 + C_4 \\
    C_4 &= 0
\end{align}


Voor snede I - tussen $[0 < x < 5]$ geldt;

Bij oplegging $A$ geldt;

$x=0, \dfrac{dv}{dx}=0$

\begin{align}
    EI\dfrac{dv}{dx} &= \dfrac{1}{2}C_1x^2 + C_2x + C_3 \\
    EI \cdot 0 &= \dfrac{1}{2}C_1 \cdot 0^2 + C_2 \cdot 0 + C_3 \\
    EI \cdot 0 &= 0 +0 + C_3\\
    C_3 &= 0
\end{align}

Voor snede I - tussen $[0 < x < 5]$ geldt;

Onder de punt belasting $F$ geldt;

$x=\dfrac{1}{2}L, \dfrac{dv}{dx}=0$

\begin{align}
    EI\dfrac{dv}{dx} &= \dfrac{1}{2}C_1x^2 + C_2x + C_3 \\
    EI \cdot 0 &= \dfrac{1}{2}C_1 \cdot (\dfrac{1}{2}L)^2 + C_2 \cdot \dfrac{1}{2}L + C_3 \\
    EI \cdot 0 &= \dfrac{1}{2}C_1 \cdot (\dfrac{1}{2}L)^2 + C_2 \cdot \dfrac{1}{2}L + 0 \\
    EI \cdot 0 &= \dfrac{1}{2}C_1 \cdot (\dfrac{1}{2}L)^2 + C_2 \cdot \dfrac{1}{2}L \\
    0 &= \dfrac{1}{2}C_1 \cdot \dfrac{1}{4}L^2 + C_2 \cdot \dfrac{1}{2}L \\
    C_2 \cdot \dfrac{1}{2}L &= - \dfrac{1}{8}C_1 \cdot L^2   \\
    C_2 &= - \dfrac{1}{4}C_1 \cdot L   \\
\end{align}

Voor snede I - tussen $[0 < x < 5]$ geldt;

Vanwege symmetrie geldt;

\begin{align}
EI\dfrac{d^3v}{dx^3} &= C_1 = \dfrac{1}{2}F
\end{align}


$C_1$ invullen geeft $C_2$:

\begin{align}
    C_2 &= - \dfrac{1}{4}C_1 \cdot L \\
    C_2 &= - \dfrac{1}{4} (\dfrac{1}{2}F) \cdot L  \\
 C_2 &= -\dfrac{1}{8}FL \\
\end{align}

De waardes voor $C_1$ en $C_2$ invullen geeft:

Voor snede I - tussen $[0 < x < 5]$ geldt;


\begin{align}
EI\dfrac{d^4v}{dx^4} &= 0 \\
EI\dfrac{d^3v}{dx^3} &= \dfrac{1}{2}F \\
EI\dfrac{d^2v}{dx^2} &= \dfrac{1}{2}Fx -\dfrac{1}{8}FL \\
EI\dfrac{dv}{dx} &= \dfrac{1}{4}Fx^2 -\dfrac{1}{8}FLx  \\
EIv &= \dfrac{1}{12}Fx^3 + \dfrac{1}{16}FLx^2 
\end{align}


Voor snede II - tussen $[5 < x < 10]$ geldt;


\begin{align}
EI\dfrac{d^4v}{dx^4} &= 0 \\
EI\dfrac{d^3v}{dx^3} &= C_5 \\
EI\dfrac{d^2v}{dx^2} &= C_5x + C_6 \\
EI\dfrac{dv}{dx} &= \dfrac{1}{2}C_5x^2 + C_6x + C_7 \\
EIv &= \dfrac{1}{6}C_5x^3 + \dfrac{1}{2}C_6x^2 + C_7x +C_8
\end{align}


Bij oplegging $B$ geldt;

$x=L, \dfrac{dv}{dx}=0$

\begin{align}
    EI\dfrac{dv}{dx} &= \dfrac{1}{2}C_5x^2 + C_6x + C_7 \\
    EI \cdot 0 &= \dfrac{1}{2}C_5 \cdot L^2 + C_6 \cdot L + C_7 \\
    EI \cdot 0 &= 0 +0 + C_3\\
    C_3 &= 0
\end{align}

Bij oplegging $B$ geldt;

$x=L, v=0$


\begin{align}
    EIv &= \dfrac{1}{6}C_5x^3 + \dfrac{1}{2}C_5x^2 + C_6x +C_7 \\
    EI\cdot 0 &= \dfrac{1}{6}C_1 \cdot L^3 + \dfrac{1}{2}C_2 \cdot L^2 + 0 \cdot L + 0  \\
    EI\cdot 0 &= \dfrac{1}{6}C_1 \cdot L^3 + \dfrac{1}{2}(- C_1 \cdot L) \cdot L^2 \\
    EI\cdot 0 &= \dfrac{1}{6}C_1 \cdot L^3 - \dfrac{1}{2}C_1 \cdot L^3 \\
    0 &= \dfrac{1}{6}C_1 \cdot L^3 - \dfrac{1}{2}C_1 \cdot L^3 \\ 
    \dfrac{1}{3}C_1 \cdot L^3  &= \dfrac{5}{24}q \cdot L^4 \\
    C_1 &= \dfrac{15}{24}q \cdot L = \dfrac{5}{8}q \cdot L
\end{align}

De functie voor de dwarskracht is bekend en kan nu getekend worden;

Voor snede I - tussen $[0 < x < 5]$ geldt;

\begin{align}
    V = EI\dfrac{d^3v}{dx^3} &= -qx + \dfrac{5}{8}ql \\
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 5     # Locatie van het koppel (m)
F = 10        # Punt belasting (kN, naar beneden)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
V_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        V_values[i] = (1 / 2) * F 
    else:  # Rechts van het koppel
        V_values[i] = -(1/2) * F  # Sprong in moment door koppel

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, V_values, label="Dwarskrachtverloop V(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk (m)")
plt.ylabel("Drarskracht V(x) [kN]")
plt.title("Dwarskrachtenlijn van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

De functie voor het buigendmoment is bekend en kan nu getekend worden;

\begin{align}
    M = EI\dfrac{d^2v}{dx^2} &= \dfrac{1}{2}Fx -\dfrac{1}{8}FL \\
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 5     # Locatie van het koppel (m)
F = 10
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
M_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        M_values[i] =  (1 / 2) * F * x - (1 / 8) * F * L
    else:  # Rechts van het koppel
        M_values[i] = (1 / 2) * F * (10-x) - (1 / 8) * F * L

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, M_values, label="Momentverloop M(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk (m)")
plt.ylabel("Moment M(x) [kNm]")
plt.title("Momentenlijn van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

De functie van de helling is bekend en kan nu getekend worden;


\begin{align}
    EI\dfrac{dv}{dx} &= \dfrac{1}{4}Fx^2 -\dfrac{1}{8}FLx 
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 5     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
F = 10
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
EIv_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        EIv_values[i] = (1/4) * F * x**2 - (1/8) * F * L * x
    else:  # Rechts van het koppel
        EIv_values[i] = -(1/4) * F * (10-x)**2 + (1/8) * F * L * (10-x)

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, EIv_values, label="EIv(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk [m]")
plt.ylabel("EIv(x) [m]")
plt.title("Helling van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

De functie van de elastischelijn (doorbuigingslijn) is bekend en kan nu getekend worden;

\begin{align}
EI\dfrac{d^4v}{dx^4} &= 0 \\
EI\dfrac{d^3v}{dx^3} &= \dfrac{1}{2}F \\
EI\dfrac{d^2v}{dx^2} &= \dfrac{1}{2}Fx -\dfrac{1}{8}FL \\
EI\dfrac{dv}{dx} &= \dfrac{1}{4}Fx^2 -\dfrac{1}{8}FLx  \\
EIv &= \dfrac{1}{12}Fx^3 + \dfrac{1}{16}FLx^2 
\end{align}

\begin{align}
E   Iv &= \dfrac{1}{12}Fx^3 + \dfrac{1}{16}FLx^2 
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 5     # Locatie van het koppel (m)
F = 10
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
EIv_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        EIv_values[i] = (1/12) * F * x**3 + (1/16) * F * L * x**2
    else:  # Rechts van het koppel
        EIv_values[i] = -(1/12) * F * (10-x)**3 - (1/16) * F * L * (10-x)**2

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, EIv_values, label="EIv(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk [m]")
plt.ylabel("EIv(x) [m]")
plt.title("Doorbuiging van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

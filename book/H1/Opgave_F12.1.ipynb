De functie van het buigendmoment kan gevonden worden door een snede te maken. 

\begin{align}
 M = 30
\end{align}

De functie voor het buigendmoment is bekend en kan nu getekend worden;

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 3       # Lengte van de balk (m)
x_C = 10     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 0   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
M_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        M_values[i] = 30
    else:  # Rechts van het koppel
        0

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, M_values, label="Momentverloop M(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(3, color='black', linewidth=2, label="Oplegging links (A)")
#plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk (m)")
plt.ylabel("Moment M(x) [kNm]")
plt.title("Momentenlijn van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

De functie voor de dwarskracht kan gevonden worden door de functie van het buigendmoment te differentiëren.

\begin{align}
 V = 0
\end{align}

De functie voor de dwarskracht is bekend en kan nu getekend worden;

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 3       # Lengte van de balk (m)
x_C = 20     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
V_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        V_values[i] = 0
    else:  # Rechts van het koppel
        V_values[i] = -10 * x + 20  # Sprong in moment door koppel

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, V_values, label="Dwarskrachtverloop V(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk (m)")
plt.ylabel("Drarskracht V(x) [kN]")
plt.title("Dwarskrachtenlijn van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

De functie voor de elastische lijn (doorbuigings) kan gevonden worden door gebruik te maken van;

$EI \dfrac{d^2v}{dx^2}= M$ en deze twee keer te integreren.


\begin{align}
EI\dfrac{d^2v}{dx^2} &= M \\
EI\dfrac{dv}{dx} &= Mx + C_1 \\
EIv &= \dfrac{1}{2}Mx^2 + C_1x + C_2\\
\end{align}


De integratie constanten $C_1$ en $C_2$ kunnen gevonden worden door gebruik te maken van de volgende randvoorwaardes.

Bij oplegging $B$ geldt;

$x=L, v'=0$


\begin{align}
EI\dfrac{dv}{dx} &= Mx + C_1 \\
EI\cdot 0 &= M \cdot L + C_1 \\
0 &= 0 + C_1 \\
 C_1 &= -ML
\end{align}


Bij oplegging $B$ geldt;

$x=L, v=0$


\begin{align}
EIv &= \dfrac{1}{2}Mx^2 + C_1x + C_2\\
EI\cdot 0 &=  \dfrac{1}{2}ML^2 - ML\cdot L + C_2\\
 0 &=  -\dfrac{1}{2}ML^2  + C_2\\
 C_2 &= \dfrac{1}{2}ML^2 
\end{align}


De waardes voor $C_1$ en $C_2$ invullen geeft:


\begin{align}
EIv &= \dfrac{1}{2}Mx^2 + C_1x + C_2\\
EIv &= \dfrac{1}{2}Mx^2 - MLx + \dfrac{1}{2}ML^2 \\
\end{align}


De elastischelijn (doorbuigingslijn) kan nu getekend worden;

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 3       # Lengte van de balk (m)
x_C = 11     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 30   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
EIv_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        EIv_values[i] = ( (1/2) * M_C * x**2 -  M_C * L * x + (1/2) * M_C * L**2  ) * 1
    else:  # Rechts van het koppel
        EIv_values[i] = 0

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, EIv_values, label="EIv(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
#plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
plt.axvline(L, color='black', linewidth=2, label="Oplegging rechts (B)")
#plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk [m]")
plt.ylabel("EIv(x) [m]")
plt.title("Doorbuiging van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```


De afgeleide van de elastische lijn is de hellingsfunctie. De hellingsfunctie is bekend en kan getekend worden.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegeven parameters
L = 10       # Lengte van de balk (m)
x_C = 11     # Locatie van het koppel (m)
q = 10     # Verdeelde belasting (kN/m, naar beneden)
M_C = 160   # Koppel (kNm, met de klok mee)
R_A = 20    # Reactiekracht bij A (kN)
R_B = 60    # Reactiekracht bij B (kN)

# Discretisatie van de balk
x_values = np.linspace(0, L, 500)
EIv_values = np.zeros_like(x_values)

# Momentverloop berekenen
for i, x in enumerate(x_values):
    if x < x_C:  # Links van het koppel
        EIv_values[i] = (-(1/6) * q * x**3 + (1/2) * q * L * x**2 - (1/2) * q * L**2 * x ) * 1
    else:  # Rechts van het koppel
        EIv_values[i] = 0

# Plot de momentenlijn
plt.figure(figsize=(8, 5))
plt.plot(x_values, EIv_values, label="EIv(x)", linewidth=2, color='blue')
plt.axhline(0, color='black', linewidth=0.8)
#plt.axvline(x_C, color='blue', linestyle='dashed', linewidth=1, label="Koppel bij x=6m")
plt.axvline(0, color='black', linewidth=2, label="Oplegging links (A)")
#plt.axvline(L, color='black', linestyle='dashed', linewidth=0.8, label="Oplegging rechts (B)")
plt.xlabel("Positie langs de balk [m]")
plt.ylabel("EIv(x) [m]")
plt.title("Helling van de balk")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```


Vanwege de vorm is bekend dat de maximale doorbuiging, $v_{max}$ optreed ter plaatse van het einde van de balk, dus $x= L$.

$x = L$ invullen geeft $v_{c} = v_{max}$:



\begin{align}
EIv_{max} &=   - \dfrac{1}{24}qx^4 + \dfrac{1}{6}qlx^3 -\dfrac{1}{4}ql^2x^2  \\
EIv_{max} &=   - \dfrac{1}{24}ql^4 + \dfrac{1}{6}qll^3 -\dfrac{1}{4}ql^2l^2  \\
EIv_{max} &=   - \dfrac{1}{24}ql^4 + \dfrac{1}{6}ql^4 -\dfrac{1}{4}ql^4  \\
EIv_{max} &=   - \dfrac{1}{24}ql^4 + \dfrac{1}{6}ql^4 -\dfrac{1}{4}ql^4 
EIv_{max} &=   -\dfrac{1}{8}ql^4 \\
v_{max}   &=   -\dfrac{1}{8EI}ql^4 \\
\end{align}




Vanwege de vorm is bekend dat de maximale helling (rotatie), $\theta_{max}$ optreed ter plaatse van het einde van de balk, dus $x= L$.

$x = L$ invullen geeft $\theta_c =  \theta_{max} = \dfrac{dv_{max}}{dx}$:

\begin{align}
EI\dfrac{dv_{max}}{dx} &= -\dfrac{1}{2}ql^2x - \dfrac{1}{6}qx^3 + \dfrac{1}{2}qlx^2 \\
EI\dfrac{dv_{max}}{dx} &= -\dfrac{1}{2}ql^2l - \dfrac{1}{6}ql^3 + \dfrac{1}{2}qll^2 \\
EI\dfrac{dv_{max}}{dx} &= -\dfrac{1}{2}ql^3 - \dfrac{1}{6}ql^3 + \dfrac{1}{2}ql^3 \\
EI\dfrac{dv_{max}}{dx} &=  - \dfrac{1}{6}ql^3  \\
\dfrac{dv_{max}}{dx} &=  - \dfrac{1}{6EI}ql^3  \\
\end{align}


Dus,

\begin{align}
\theta_{max } = - \dfrac{1}{6EI}ql^3  \\
\end{align}
